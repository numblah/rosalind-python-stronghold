```python
import pandas as pd
with open('rosalind_grph.txt', 'r') as f:
    data = f.readlines()
    
data=''.join(data)
```

```python
def FASTA(data):
    import pandas as pd
    
    split_data=data.split('>')

    del split_data[0]
    split_data

    for index, element in enumerate(split_data):
        split_data[index]=element.split('\n',1)

    data_df=pd.DataFrame (split_data, columns=['ID','seq']) 
    data_df['seq'].replace(regex=True, inplace=True, to_replace='\n', value=r'')
    return(data_df)
```

```python
data_df=FASTA(data)
data_df
```
